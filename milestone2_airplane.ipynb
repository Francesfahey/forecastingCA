```python

# import the required libs
#!pip install statsmodels
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


from statsmodels.tsa.seasonal import seasonal_decompose, STL
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller
```

```python
data_url='https://raw.githubusercontent.com/Francesfahey/forecastingCA/main/data/AirPassengers2.csv'
df = pd.read_csv(data_url)
# show top records in the df
df.tail()
```

```python

fig = plt.figure()
plt.plot(df.Month, df.Passengers)
plt.title('Airplane passengers')
plt.xlabel('Month')
plt.ylabel('no. of passengers')

plt.xticks(rotation = 100, ha='right')

fig.autofmt_xdate()
plt.tight_layout()
plt.show()
plt.savefig('airplane.png', dpi=300)
```

```python
ADF_result = adfuller(df['Passengers'])

print(f'ADF Statistic: {ADF_result[0]}')
print(f'p-value: {ADF_result[1]}')
```

```python
plot_acf(df['Passengers'], lags=20);

plt.tight_layout()
plt.savefig('fig_passengers.png', dpi=300)
```

```python
Passengers_diff = np.diff(df['Passengers'], n=5)
print(Passengers_diff)
```

```python
ADF_result = adfuller(Passengers_diff)

print(f'ADF Statistic: {ADF_result[0]}')
print(f'p-value: {ADF_result[1]}')
```

```python
plot_acf(Passengers_diff, lags=20);


plt.tight_layout()
plt.savefig('fig_passengers.png', dpi=300)
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Create sample data
Passengers_diff = np.array([1, 2, 3, 4, 5])

# Create MA process
ma21d = np.array([1, 2, 3])
MA2_process = ArmaProcess(Passengers_diff, ma21d)

# Plot time series data
fig, ax = plt.subplots()
ax.plot(Passengers_diff)
ax.set_xlabel('Timesteps')
ax.set_ylabel('Value')
fig.autofmt_xdate()
plt.tight_layout()

# Plot ACF function
plot_acf(Passengers_diff, lags=30)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(MA2_process)
ax.set_xlabel('Timesteps')
ax.set_ylabel('Value')

fig.autofmt_xdate()
plt.tight_layout()


# ACF function
plot_acf(MA2_process, lags=30);
```
